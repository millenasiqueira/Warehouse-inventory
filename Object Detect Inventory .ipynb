Train the model

```python
# Train the model
results = model.train(
    data=f"/content/Inventory--1/data.yaml",
    epochs=100,  # You can adjust the number of epochs
    imgsz=640,   # Image size
    batch=16,    # Batch size
    name='Inventory',  # Name for the experiment
    patience=5 # Stop early if no improvements 5 epochs in a row
)
```

To see

```python
from ultralytics import YOLO
from PIL import Image

# Load the trained model (replace 'Inventory/weights/best.pt' with your actual model path if different)
model = YOLO('/content/yolov5/runs/detect/Inventory4/weights/best.pt')

# Load an image for detection
image_path = '/content/WhatsApp Image 2024-10-28 at 18.15.42.jpeg'  # Replace with the path to your image
image = Image.open(image_path)

# Run inference

results = model.predict(source=image_path, imgsz=640)  # Predict on the image

# Display results and print detected objects
for result in results:
    result.plot()  # Plot bounding boxes on the image
    result.show()  # Display the image (this will work if running locally in a Jupyter environment)

    # Print detected objects and their confidence scores
    for detection in result.boxes:
        class_id = int(detection.cls)  # Class ID of detected object
        confidence = detection.conf  # Confidence score of detection
        print(f"Detected {model.names[class_id]} with confidence {confidence:.2f}")
```
